```python
import pandas as pd

# original untouched data
df_all = pd.read_csv("pqm_extract.csv")

# keep only plans with a slash in PlanId
df = df_all[
    df_all["PlanId"]
    .fillna("")
    .str.contains("/", regex=False)
].copy()

print(f"Original rows : {len(df_all):,}")
print(f"Rows with '/' : {len(df):,}")

df.head()
```

```python
df_all.loc[
    ~df_all["PlanId"].fillna("").str.contains("/", regex=False),
    ["PatientId", "CourseId", "PlanId"]
].drop_duplicates().sort_values(["PatientId", "PlanId"])
```
